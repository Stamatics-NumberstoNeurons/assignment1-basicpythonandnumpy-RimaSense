### Exercise 15 — Vectorize a Function

**Question:** Convert the scalar function `maxx` below to work on two arrays.


```python
# Input — do not modify
def maxx(x, y):
    """Get the maximum of two items"""
    if x >= y:
        return x
    else:
        return y

a = np.array([5, 7, 9, 8, 6, 4, 5])
b = np.array([6, 3, 4, 8, 9, 7, 1])
```

```python
# Write your code here:
vectorized_maxx = np.vectorize(maxx)
result = vectorized_maxx(a, b)
print(result)
```

### Exercise 40 — Numeric to Categorical

**Question:** Bin the petal length (3rd column) of `iris_2d` to form a text array using these rules:
- `< 3` → `'small'`
- `3–5` → `'medium'`
- `>= 5` → `'large'`

*# Make sure you have run Exercise 27 first.*


```python
# Write your code here:
petal_length = iris_2d[:, 2]
bin_edges = [3, 5]
bin_indexes = np.digitize(petal_length, bins=bin_edges)
label_map = np.array(['small', 'medium', 'large'])
petal_categories = label_map[bin_indexes]
print(petal_categories)

#new 2d matrix with replaced 3rd column
iris_mixed = iris_2d.astype('object')
iris_mixed[:, 2] = petal_categories
print(iris_mixed)
```

### Exercise 41 — Create New Column

**Question:** Create a new column for **volume** in `iris_2d`, where:

> volume = (π × petal_length × sepal_length²) / 3

*# Make sure you have run Exercise 27 first.*


```python
# Write your code here:
sepal_length = iris_2d[:, 0]
petal_length = iris_2d[:, 2]
volume = (np.pi * petal_length * (sepal_length ** 2)) / 3
iris_3d = np.column_stack((iris_2d, volume))
print(iris_3d)
```

### Exercise 42 — Probabilistic Sampling

**Question:** Randomly sample iris's species such that **setosa** appears **twice** as often as versicolor and virginica.

*# Make sure you have run Exercise 25 first.*


```python
# Write your code here:
species_column = iris_1d['f4']
unique_species = np.unique(species_column)
probabilities = [0.50, 0.25, 0.25]
rigged_sample = np.random.choice(unique_species, size=150, p=probabilities, replace=True)
sampled_species, counts = np.unique(rigged_sample, return_counts=True)
print("Rigged Sample Distribution:")
for name, count in zip(sampled_species, counts):
    print(f"{name}: {count}")
```
